```python
import pandas as pd

df = pd.read_csv("train.csv")
print(df.columns)
print(df.head())
```

```python
import jieba

# 取评论列
sentences = df['comment'].dropna().tolist()

# 分词
sentences_cut = [list(jieba.cut(s)) for s in sentences]

print("样本数：", len(sentences_cut))
print("示例：", sentences_cut[0])
```

```python
from gensim.models import Word2Vec

# 训练模型（Skip-Gram）
model = Word2Vec(
    sentences_cut,
    vector_size=100,
    window=5,
    min_count=2,
    sg=1   
)

print("训练完成！")
```

```python
print(model.wv['环境'])
print("向量维度：", model.wv['环境'].shape)
```

```python
print(model.wv.most_similar("好吃", topn=3))
```

```python
print("好吃 vs 美味：", model.wv.similarity("好吃", "美味"))
print("好吃 vs 蟑螂：", model.wv.similarity("好吃", "蟑螂"))
```

```python
print(model.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'], topn=1))
```
